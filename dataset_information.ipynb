```python
data_pos = dataset[dataset['COND'] == '+']
data_neg = dataset[dataset['COND'] == '-']
data_neut = dataset[dataset['COND'] == '0']
```

```python
fig, ax = plt.subplots()
ax.scatter(data_neut['ANS_VALENCE'], data_neut['ANS_AROUSAL'], marker='.', color='blue', alpha=0.2)
ax.scatter(data_pos['ANS_VALENCE'], data_pos['ANS_AROUSAL'], marker='.', color='green', alpha=0.2)
ax.scatter(data_neg['ANS_VALENCE'], data_neg['ANS_AROUSAL'], marker='.', color='red', alpha=0.2)
plt.xlabel('Valence')
plt.ylabel('Arousal')
plt.show()
```
